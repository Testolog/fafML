```python
from typing import Type
from pandas import DataFrame, read_json, Series
from pandas.io.json import json_normalize
import os
import json
import numpy as np
```

```python
data = json.loads("".join(open("./blueprints.json", 'r')))
rawDF: DataFrame = DataFrame(data)
rawDF: DataFrame = rawDF.set_index("Id")
rawDF.head()
```

```python
# t = DataFrame.from_dict(rawDF['Economy'], )
# todo find how to make it more beauty
def convert_dict_to_df(serias:Series):
    raw_normalize_data = {}
    for index, row in serias.items():
        if type(row) is dict:
            for key,value in row.items():
                if key not in raw_normalize_data:
                    raw_normalize_data[key] = [(index, value)]
                else:
                    raw_normalize_data[key].append((index,value))
    
    series_data = {}
    for key, value in raw_normalize_data.items():
        series_data[key] = Series([_r[1] for _r in value], index=[_r[0] for _r in value])
    return DataFrame(series_data)

convert_dict_to_df(rawDF['Economy'])

# rawDF['Economy'] = convert_dict_to_df(rawDF['Economy'])
# rawDF['Economy'].head()
# DataFrame(np.nan, columns=columns, index=rawDF.index)
```
